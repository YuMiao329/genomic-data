```python
# df2: 2 columns
df2 =  df[['Tumor_Sample_Barcode','Hugo_Symbol']]
df2
```

```python
# df3: no duolicates
# df4: new dataframe
df3 = df2.drop_duplicates(subset = ["Hugo_Symbol"])
column_names = df3['Hugo_Symbol']
df4 = pd.DataFrame(columns = column_names)
df4.rename(columns = {"Hugo_Symbol": "Tumor_Sample_Barcode"}, 
          inplace = True)
df4
```

```python
# for test, don't run
# df3.loc['TCGA-AA-3975-01','Hugo_Symbol']=1
# df3
```

```python

# Use df4, add data in df2(with all gene)
# 223 sample, 15980 gene mutation position

for i, gene in enumerate(df2['Hugo_Symbol']):
    row = str(df2.iloc[i, 0])
    # print('row', row)
    column = str(gene)
    # print('column',column)
    # break
    df4.loc[row,column] = 1
    # df.at[1,'Hugo_Symbol']=1
df4 = df4.fillna(0)
df4
```

```python
# df_mutation
# drop last row of nan, drop last 2 columns of 1
# df4.drop(df4.tail(1).index,inplace=True)
df_mutation = df4.iloc[:-1 , :-2]
df_mutation
```

```python
df_mutation.to_csv('mutation.csv')  
```

```python
# df5: clinical data witout na
df5 = pd.read_csv('coad_tcga_clinical_data.csv')
df5 = df5[['Sample ID','Overall Survival Status']]
df5 = df5.dropna()
df5
```

```python
# df5: mutation with Overall Survival Status
df_clinic = df_mutation

for i, status in enumerate(df5['Overall Survival Status']):
    row = str(df5.iloc[i, 0])
    # print('row', row)
    # print('column',status)
    # break
    df_clinic.loc[row,'status'] = status

df_clinic
```

```python
# df6: df_clinic drop na
df6 = df_clinic.dropna()
df6
```

```python
df6.to_csv('mutation_with_status.csv')  
```
